```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('../data/raw/titanic/train.csv')
df2 = pd.read_csv('../data/raw/titanic/test.csv')
print(df1.shape)
print(df2.shape)

```

```python
print("Train columns:", df1.columns.tolist())
print("Test columns:", df2.columns.tolist())
```

```python
print("Missing values in train:")
print(df1.isnull().sum())
print("\nMissing values in test:")
print(df2.isnull().sum())
```

```python
# Data types
print("\nData types in train:")
print(df1.dtypes)
```

```python
print("\nSummary statistics:")
print(df1.describe(include='all'))
```

```python
sns.countplot(x='Survived', data=df1)
plt.title('Survival Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,6))
numeric_df = df1.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

```python
categorical_features = ['Pclass', 'Sex', 'Embarked']
for feature in categorical_features:
    sns.countplot(x=feature, hue='Survived', data=df1)
    plt.title(f'{feature} vs Survival')
    plt.show()

# Age distribution
sns.histplot(df1['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()

# Fare distribution
sns.histplot(df1['Fare'], bins=30, kde=True)
plt.title('Fare Distribution')
plt.show()
```
